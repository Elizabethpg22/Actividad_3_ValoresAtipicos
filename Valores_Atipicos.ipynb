```python
d1 =pd.read_csv('VentasTotales_Limpio.csv')
d1= d1.drop(['Unnamed: 0'],axis=1)
```

###Columna ventas_precios_corrientes

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize = (5,3))
plt.boxplot(d1['ventas_precios_corrientes'])
plt.title('Outliers de ventas_precios_corrientes')
plt.show() #Dibujamos el diagrama
```

```python
#Metodo aplicado Cuartiles. Encunetro cuartiles 0.25 y 0.75
y = d1['ventas_precios_corrientes']

percentile25 = y.quantile(0.25) #Q1
percentile75 = y.quantile(0.75) #Q3

iqr = percentile75 -percentile25
Limite_Superior_iqr = percentile75 + 1.5*iqr
Limite_Inferior_iqr = percentile25 - 1.5*iqr
print('Limite superior permitido', Limite_Superior_iqr)
print('Limite Inferior  permitido', Limite_Inferior_iqr)
```

```python
#Encontramos Outliers
outliers_iqr = d1[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtener datos Limpios
data_clean_ventas_precios_corrientes = d1[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
```

```python
#Realizamosla comprobacion de valores atipicos
fig = plt.figure(figsize = (5,3))
plt.boxplot(data_clean_ventas_precios_corrientes['ventas_precios_corrientes'])
plt.title('Outliers de ventas_precios_corrientes')
plt.show() #Dibujamos el diagrama
```

```python
data_clean_ventas_precios_corrientes.to_csv("data clean_ventas precios corrientes.csv")
from google.colab import files
files.download("data clean_ventas precios corrientes.csv")
```
